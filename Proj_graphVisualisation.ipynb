```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('SalesOutput.txt', delimiter="\t", header= None, names =['Store', 'Revenue'])
ax = data[['Revenue']].plot(kind='bar', title ="Total Revenue Analysis", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Store", fontsize=12)
ax.set_ylabel("Count of Transactions (", fontsize=12)
ax.set_ylim(9800000,10200000)
plt.show()
```

```python
data = pd.read_csv('PaymentOutput.txt', delimiter="\t", header= None, names =['Store', 'Cash', 'Credit'])
ax = data[['Cash']].plot(kind='bar', title ="Payment Analysis", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Store", fontsize=12)
ax.set_ylabel("Count of Transactions", fontsize=12)
ax.set_ylim(7500,8500)
plt.show()
```

```python
data = pd.read_csv('PaymentOutput.txt', delimiter="\t", header= None, names =['Store', 'Cash', 'Credit'])
ax = data[['Credit']].plot(kind='bar', title ="Payment Analysis", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Store", fontsize=12)
ax.set_ylabel("Count of Transactions", fontsize=12)
ax.set_ylim(30000,35000)
plt.show()
```
